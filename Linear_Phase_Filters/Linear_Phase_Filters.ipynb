# Removing the phase information

As demonstrated in the cell below, the phase information of a signal is important to determine the shape and timing of the sound wave. If we completely remove the phase information (by applying the absolute value to the Fourier Transform and restoring the sound from that), we completely loose the sharp start of the tone, but keep the same pitch.

```python
lpf.remove_phase()
```
